```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = 'data/reduced'
```

```python
data = {}
data["M29B"] = np.load(f'{data_path}/M29B.npy')
data["M29V"] = np.load(f'{data_path}/M29G.npy')
```

```python
error = {}
error["M29B"] = np.load(f'{data_path}/M29B_error.npy')
error["M29V"] = np.load(f'{data_path}/M29G_error.npy')[0]
```

```python
for key, image in data.items():
    lo,up= np.percentile(image,1),np.percentile(image,99)
    plt.imshow(image, cmap='gray', origin='lower', clim=(lo,up))
    plt.title(f'{key}')
    plt.show()
```

```python
for key, image in error.items():
    lo,up= np.percentile(image,1),np.percentile(image,99)
    plt.imshow(image, cmap='gray', origin='lower', clim=(lo,up))
    plt.title(f'{key} error')
    plt.show()
```

```python
images = [data["M29B"], data["M29V"]]

result = np.median(np.stack(images, axis=0), axis=0)

lo,up= np.percentile(result,1),np.percentile(result,99)
plt.imshow(result, cmap='gray', origin='lower', clim=(lo,up))
plt.title(f'BV stacked')
plt.show()
```
